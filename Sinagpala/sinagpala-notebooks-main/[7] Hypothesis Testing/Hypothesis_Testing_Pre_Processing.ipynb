```python
import pandas as pd
import geopandas as gpd
```

## Import file

```python
gsa_pvout = gpd.read_file('./PVOUT_yearly_monthly.geojson')
```

## Remove unneccessary columns

```python
essential_columns = ['id','geometry'] #no address
pvout_columns = ['PVOUT_yearly_mean', 'PVOUT_01_mean', 'PVOUT_02_mean', 'PVOUT_03_mean', 
                 'PVOUT_04_mean', 'PVOUT_05_mean', 'PVOUT_06_mean', 'PVOUT_07_mean', 
                 'PVOUT_08_mean', 'PVOUT_09_mean', 'PVOUT_10_mean', 'PVOUT_11_mean', 
                 'PVOUT_12_mean']
```

```python
essential_columns += pvout_columns
essential_columns
```

```python
pvout_df = gsa_pvout[essential_columns]
pvout_df = pvout_df[pvout_df['geometry'].apply(lambda x : x.type!='Point' )]

pvout_df.rename(columns = {'PVOUT_yearly_mean':'PVOUT_yearly', 
                           'PVOUT_01_mean':'PVOUT_month_1',
                           'PVOUT_02_mean':'PVOUT_month_2',
                           'PVOUT_03_mean':'PVOUT_month_3',
                           'PVOUT_04_mean':'PVOUT_month_4',
                           'PVOUT_05_mean':'PVOUT_month_5',
                           'PVOUT_06_mean':'PVOUT_month_6',
                           'PVOUT_07_mean':'PVOUT_month_7',
                           'PVOUT_08_mean':'PVOUT_month_8',
                           'PVOUT_09_mean':'PVOUT_month_9',
                           'PVOUT_10_mean':'PVOUT_month_10',
                           'PVOUT_11_mean':'PVOUT_month_11',
                           'PVOUT_12_mean':'PVOUT_month_12',}, 
                inplace = True)
pvout_df = pvout_df.drop(columns=['geometry'])

pvout_df.head()
```

## import solar potential df

```python
solar_potential = gpd.read_file('../[5] Solar Energy Potential Calculation/solar_potential_2007_2018_flatsouth.geojson')
```

```python
solar_potential.head()
```

## Compute kWp

```python
# 1kWp = four 250w solar panels (https://solvoltaics.com/calculate-kwp-solar-panel/)

solar_potential['kWp'] = solar_potential['num_panels'] / 4
```

```python
solar_potential.head()
```

## Merge PVOUT df and solar potential df

```python
new_df = solar_potential.merge(pvout_df, left_on='id', right_on='id')
new_df = new_df[new_df['num_panels'] > -999] # exclude buildings with no data
```

```python
print(new_df.columns)
```

```python
new_df = new_df[['id', 'address', 'num_panels', 'panel_area', 'geometry', 'kWp', 
                 'total_kwh',  
                 'avg_month_1', 'avg_month_2', 'avg_month_3', 'avg_month_4',
                 'avg_month_5', 'avg_month_6', 'avg_month_7', 'avg_month_8',
                 'avg_month_9', 'avg_month_10', 'avg_month_11', 'avg_month_12',
                 'PVOUT_yearly',
                 'PVOUT_month_1', 'PVOUT_month_2', 'PVOUT_month_3', 'PVOUT_month_4', 
                 'PVOUT_month_5', 'PVOUT_month_6', 'PVOUT_month_7', 'PVOUT_month_8', 
                 'PVOUT_month_9', 'PVOUT_month_10', 'PVOUT_month_11', 'PVOUT_month_12']]
```

```python
pvout_columns = ['PVOUT_yearly',
                 'PVOUT_month_1', 'PVOUT_month_2', 'PVOUT_month_3', 'PVOUT_month_4', 
                 'PVOUT_month_5', 'PVOUT_month_6', 'PVOUT_month_7', 'PVOUT_month_8', 
                 'PVOUT_month_9', 'PVOUT_month_10', 'PVOUT_month_11', 'PVOUT_month_12']

for column in pvout_columns:
  new_df[column] = new_df[column] * new_df['kWp'] # convert kWh/kWp value of PVOUT to kWh
```
